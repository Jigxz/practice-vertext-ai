```python
from langchain_google_vertexai import ChatVertexAI
from langchain_core.messages import HumanMessage, AIMessage

# Initialize the ChatVertexAI model
# You might need to specify project and location depending on your setup
chat = ChatVertexAI(model="gemini-2.5-flash",project=VERTEXAI_PROJECT) 

# Initialize an empty list to store the chat history
chat_history = []

def get_chat_response(user_input: str):
    """
    Sends a user message to ChatVertexAI with history and returns the model's response.
    """
    global chat_history

    # Add the current user message to the history
    chat_history.append(HumanMessage(content=user_input))

    # Invoke the model with the entire chat history
    response = chat.invoke(chat_history)

    # Add the model's response to the history
    chat_history.append(AIMessage(content=response.content))
    
    return response.content

# Example usage
print(f"User: Hello")
response1 = get_chat_response("Hello")
print(f"AI: {response1}\n")

print(f"User: My name is John!")
response2 = get_chat_response("My name is John!")
print(f"AI: {response2}\n")

print(f"User: What is my name?")
response2 = get_chat_response("What is my name?")
print(f"AI: {response2}\n")

print(f"User: And what did you say before that?")
response3 = get_chat_response("And what did you say before that?")
print(f"AI: {response3}\n")
```
